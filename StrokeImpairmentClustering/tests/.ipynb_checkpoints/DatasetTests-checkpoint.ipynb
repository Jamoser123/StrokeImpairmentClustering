```python
import matplotlib.pyplot as plt

# Set global font size and DPI
plt.rcParams['figure.dpi'] = 300

def plot_simple_time_series(data, title):
    # Create a 2D plot
    plt.figure(figsize=(8, 4))
    
    # Define axis labels for the legend
    axis_labels = ['x-axis', 'y-axis', 'z-axis']
    
    # Plot each axis (assuming data.shape[1] == 3 for x, y, z)
    for i in range(data.shape[1]):
        plt.plot(data[:, i], label=axis_labels[i])
    
    plt.title(title, fontsize=20)
    plt.legend()
    plt.xlabel("Time [s]", fontsize=20)
    plt.ylabel("Angular Velocity [Â°/s]", fontsize=20)
    plt.tight_layout()
    plt.savefig('/cluster/home/jamoser/plots/acceleration.png')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import torch
import numpy as np

# Set global font size and DPI
plt.rcParams['figure.dpi'] = 300

# Function to apply individual augmentations
def apply_noise(data, noise_level=0.1):
    noise = torch.randn_like(data) * noise_level
    return data + noise

def apply_mask(data, mask_ratio=0.2):
    masked_data = data.clone()
    seq_len = data.shape[0]
    mask_length = int(seq_len * mask_ratio)
    if mask_length > 0 and (seq_len - mask_length) > 0:
        start_idx = torch.randint(0, seq_len - mask_length, (1,)).item()
        masked_data[start_idx:start_idx + mask_length] = 0
    return masked_data

def apply_shuffling(data, n_segments=4):
    seq_len, num_variables = data.shape
    slice_size = seq_len // n_segments
    num_slices = n_segments
    sliced_data = data[:num_slices * slice_size].reshape(num_slices, slice_size, num_variables)
    permuted_indices = torch.randperm(num_slices)
    shuffled_slices = sliced_data[permuted_indices, :, :]
    shuffled_data = shuffled_slices.reshape(-1, num_variables)
    return shuffled_data

# Plot function for a single augmentation
def plot_single(data, title, axis_label='x-axis', save_path=None):
    plt.figure(figsize=(8, 4))
    plt.plot(data[:, 0], label=axis_label)  # Plot only the x-axis acceleration
    plt.title(title, fontsize=20)
    plt.xlabel("Time [s]", fontsize=20)
    plt.ylabel("Acceleration [m/sÂ²]", fontsize=20)
    # plt.legend()
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path)
    plt.show()
```

```python
# Example usage
if __name__ == "__main__":
    # Generate dummy data for x, y, z accelerations
    # seq_len = 100
    # num_variables = 3  # x, y, z axes
    data = torch.tensor(ts_numpy[1000:1600, :], dtype=torch.float32)  # Random dummy data

    # 1. Original Data
    plot_single(data, title="Original X-Axis Acceleration", axis_label="x-axis",
                save_path="/cluster/home/jamoser/plots/original.png")

    # 2. Noise Augmentation
    noisy_data = apply_noise(data, noise_level=0.1)
    plot_single(noisy_data, title="X-Axis Acceleration with Noise", axis_label="x-axis",
                save_path="/cluster/home/jamoser/plots/noise.png")

    # 3. Masking Augmentation
    masked_data = apply_mask(data, mask_ratio=0.15)
    plot_single(masked_data, title="X-Axis Acceleration with Masking", axis_label="x-axis",
                save_path="/cluster/home/jamoser/plots/masking.png")

    # 4. Shuffling Augmentation
    shuffled_data = apply_shuffling(data, n_segments=3)
    plot_single(shuffled_data, title="X-Axis Acceleration with Shuffling", axis_label="x-axis",
                save_path="/cluster/home/jamoser/plots/shuffling.png")
```
